```python
import dragon
import multiprocessing as mp
import llm_backend as lm
import time
import telemetry as tm
```

```python
mp.set_start_method("dragon")
```

```python
num_workers=4
num_readers=2
inf_workers = lm.InfWorkers(num_workers, num_readers)
telem_workers = tm.TelemWorkers()
```

```python
telem_workers.start()
```

```python
inf_workers.start()
```

```python
inf_workers.prompt('I love summer as well. I can do my favorite outdoor activities.')
```

```python
prompt_list = ['Have you heard of HPE?',
'What do you know about HPE?',
'What do you know about the US National Labs?',
'What is your favorite color?',
'My favorite color is yellow.',
'What is your favorite season?',
'I really like spring. There are so many things to do!',
'What is your favorite thing to do on a rainy day?',
'When it is raining I like to go outside anyways!',
'I love to spend time outdoors.',
'What are your favorite outdoor activities?',
'What do you like most about surfing?',
'Where is your favorite place to surf?', 
'Hawaii is awesome! What islands have you visited?',
'Have you visited many national parks?']

num_times_prompt_list=0
while True: 
    for input_prompt in prompt_list:
        inf_workers.prompt(input_prompt)
    time.sleep(1)
    num_times_prompt_list += 1
    print(num_times_prompt_list, flush=True)
    if num_times_prompt_list > 7:
        break
```

```python
inf_workers.close()
```
