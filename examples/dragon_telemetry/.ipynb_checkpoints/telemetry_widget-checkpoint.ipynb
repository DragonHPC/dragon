```python
import dragon
```

```python
import ipywidgets
from ipywidgets import interact, interactive, fixed, interact_manual
```

```python
import dragon.telemetry.widget as dtw
```

```python
import dragon.telemetry.telemetry as dt
```

```python
widget = dtw.DragonTelemetryWidget()
```

```python
dragon_telemetry = dt.Telemetry()
```

```python
dragon_telemetry.start()
```

```python
widget.connect()
```

```python
out = ipywidgets.Output()

@out.capture(clear_output=True)
def render(metric_name, percentage_window):
    display(widget.render(metric_name, percentage_window))

interact(render, metric_name=list(widget.get_metrics()), percentage_window=widget.custom_slider())
```

```python
widget.get_metrics()
```

```python
dragon_telemetry.shutdown()
```
